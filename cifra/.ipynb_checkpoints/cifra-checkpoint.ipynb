#### Проверка расчета эффективности обогащения

```python
#Проверьте, что эффективность обогащения рассчитана правильно. Вычислите её на обучающей выборке для признака 
#rougher.output.recovery. Найдите MAE между вашими расчётами и значением признака. Опишите выводы.
def recovery_check (row):
    c=row['rougher.output.concentrate_au']
    f=row['rougher.input.feed_au']
    t=row['rougher.output.tail_au']
    try:
        recovery=(100*c*(f-t))/(f*(c-t))
        return recovery
    except:
        return 0
```

```python
gold_recovery_train['recovery_check']=gold_recovery_train.apply(lambda x: recovery_check(x), axis=1)
```

```python
mae=(gold_recovery_train['rougher.output.recovery']-gold_recovery_train['recovery_check']).abs().mean()
mae
```

**Вывод**    
Значение MAE между исходной и рассчетной эффективноcтями обогащения очень маленькое. Ошибка не существенна. Изначально все посчитано верно.

#### Исследуование суммарной концентрации всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах

```python
#2.3. Исследуйте суммарную концентрацию всех веществ на разных стадиях: в сырье, в черновом и финальном концентратах. 
#Заметили аномальные значения в суммарном распределении или нет? Если они есть, стоит ли их удалять из обеих выборок? 
#Опишите выводы и удалите аномалии.
def concentrate_sum (df):
    
    df['rougher.input.feed_sum']=df['rougher.input.feed_au']+df[
    'rougher.input.feed_ag']+df['rougher.input.feed_pb']+df['rougher.input.feed_sol']

    df['rougher.output.concentrate_sum']=df['rougher.output.concentrate_au']+df[
    'rougher.output.concentrate_ag']+df['rougher.output.concentrate_pb']+df[
    'rougher.output.concentrate_sol']

    df['primary_cleaner.output.concentrate_sum']=df['primary_cleaner.output.concentrate_au'
    ]+df['primary_cleaner.output.concentrate_ag']+df['primary_cleaner.output.concentrate_pb'
    ]+df['primary_cleaner.output.concentrate_sol']

    df['final.output.concentrate_sum']=df['final.output.concentrate_au']+df[
    'final.output.concentrate_ag']+df['final.output.concentrate_pb']+df[
    'final.output.concentrate_sol']

concentrate_sum(gold_recovery_full)
concentrate_sum(gold_recovery_train)
```

### Модель

```python
gold_recovery_train.shape, gold_recovery_test.shape, gold_recovery_full.shape
```

```python
features_train=gold_recovery_train[gold_recovery_test.columns].drop(['date'], axis=1)
target_train=gold_recovery_train[['final.output.recovery', 'rougher.output.recovery']]
features_test=gold_recovery_test.drop(['date'], axis=1)

#Добавлю недостающие столбцы в train методом merge с gold_recovery_full
test_aux=gold_recovery_test.merge(gold_recovery_full, on='date', how='left')
target_test=test_aux[['final.output.recovery', 'rougher.output.recovery']]
```

```python
features_train.shape, target_train.shape, features_test.shape, target_test.shape
```

```python
#3.1 Напишите функцию для вычисления итоговой sMAPE.
def smape(target, predict):
    a=0
    predict=pd.DataFrame(data=predict, columns=target.columns)
    for i in range(len(target)):
        yi=target[['final.output.recovery', 'rougher.output.recovery']]
        yj=predict[['final.output.recovery', 'rougher.output.recovery']]
        y=(yi-yj).abs()/(((yi).abs()+(yj).abs())/2)
        #print(y)
    smape_rougher=y['rougher.output.recovery'].sum()*100/len(target)
    smape_final=y['final.output.recovery'].sum()*100/len(target)
    smape=0.25*smape_rougher+0.75*smape_final
    
    return smape
```

```python
my_scorer = make_scorer(smape, greater_is_better=False)
```

```python
#Дерево
df11 = pd.DataFrame()

for i in range(1, 9, 3):
    model = DecisionTreeRegressor(random_state=12345, max_depth=i)
    final_score=abs(cross_val_score(model, features_train, target_train, cv=3, scoring=my_scorer)).sum()/3
    df11.loc[i, 'smape']=final_score
        
print(df11.sort_values(by='smape'))
```

```python
#Случайный лес
df12=pd.DataFrame()
k=0
for i in range(1,13,10):
    for j in range(1,13,10):
        k+=1
        model = RandomForestRegressor(random_state=12345, max_depth=i, n_estimators=j)
        final_score=abs(cross_val_score(model, features_train, target_train, cv=3, scoring=my_scorer)).sum()/3
        
        df12.loc[k, 'max_depth']=i
        df12.loc[k, 'n_estim']=j
        df12.loc[k, 'smape']=final_score
    
df12.sort_values(by='smape')
```

```python
#Линейная регрессия
model = LinearRegression()
final_score=abs(cross_val_score(model, features_train, target_train, cv=3, scoring=my_scorer)).sum()/3
print('Средняя оценка качества модели:', final_score)
```

```python
model = RandomForestRegressor(random_state=12345, max_depth=1, n_estimators=11)
model.fit(features_train, target_train)
#Проверяю качество модели на тестовой выборке
predictions_test=model.predict(features_test)

smape(target_test, predictions_test)
```

```python
#Костантая модель
a=target_train['final.output.recovery'].median()
b=target_train['rougher.output.recovery'].median()
constant=[[a, b]]*target_test.shape[0]

smape(target_test, constant)
```

**Вывод**    
Обучила модели Дерева решений, Случайного леса, а также линейной регрессии. Оценила их качество кросс-валидацией.  
Лучшая модель, модель Случайного леса с параметрами max_depth=1 и n_estimators=11, показала результат Smape=3.22.   
При проверке на тестовой выборке Smape=6.86%. Это лучше, чем результат константной модели smape=7.02%.
