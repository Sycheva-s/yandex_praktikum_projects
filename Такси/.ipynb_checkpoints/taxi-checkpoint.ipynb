***Описание проекта***

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

*Описание данных*

Количество заказов находится в столбце '*num_orders*'

## Обзор данных

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV

from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import TimeSeriesSplit
```

```python
df=pd.read_csv(r'C:\Users\Susanna\Downloads\taxi.csv', index_col=[0], parse_dates=[0])
#df=pd.read_csv(r'/datasets/taxi.csv', index_col=[0], parse_dates=[0])

display(df.head())
df.info()
```

```python
df=df.resample('1H').sum()
```

**Вывод**  
Данные загружжены, выполнено ресемплирование по одному часу.

## Подготовка данных

```python
df.plot(figsize=(15, 5))
plt.show()
```

```python
decomposed = seasonal_decompose(df)
```

```python
plt.figure(figsize=(12, 12))
plt.subplot(311)
decomposed.trend.plot(ax=plt.gca())
plt.title('Trend')

plt.subplot(312)
decomposed.seasonal.plot(ax=plt.gca())
plt.title('Seasonality')

plt.subplot(313)
decomposed.resid.plot(ax=plt.gca())
plt.title('Residuals')
plt.tight_layout()
```

**Вывод**    
Добавила метод декомпозиции. Из результатов видно что в данных есть тренд на увеличение заказов такси.  
Сезонность рассмотрю на более маленьком отрезке времени:

```python
plt.figure(figsize=(12, 4))
decomposed.seasonal['2018-03-01':'2018-03-07'].plot(ax=plt.gca())
plt.show()
```

**Вывод**     
Видно, что есть какая-то закономерность в течении одного дня. Рассмотрю подробнее:

```python
plt.figure(figsize=(12, 4))
decomposed.seasonal['2018-03-01':'2018-03-02'].plot(ax=plt.gca())
plt.show()
```

**Вывод**    

Видно, что самая низкая загруженность находится в районе 6 утра. А пиковая нагрузка приходится на 12 часов ночи. 

```python
display(df.head())
```

```python
def make_features(data, max_lag, rolling_mean_size):
    data['dayofweek'] = data.index.dayofweek
    data['hour'] = data.index.hour
    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()


make_features(df, 40, 10)
```

```python
df=df.dropna()
display(df.head())
```

```python
features_train, features_test, target_train, target_test = train_test_split(
    df.drop('num_orders', axis=1), df.num_orders, test_size=0.1, random_state=12345, shuffle=False)
features_train.shape, features_test.shape, target_train.shape, target_test.shape
```

**Вывод**    
Для построения моделей добавила в датасет календарные признаки (год, месяц и день),
а также отстающие значения ременного ряда и скользящее среднее.  
Разбила данные на train и test. Тестовая выборка составляет 10% от исходных данных. 

## Обучение моделей

### LinearRegression

```python
model=LinearRegression()
model.fit(features_train, target_train)

print("RMSE:", mean_squared_error(model.predict(features_test), target_test)**0.5)
```

```python
def plotting(target_test, predicted):
    predicted=pd.DataFrame(data=predicted, index=target_test.index)
    plt.figure(figsize=(17,5))
    plt.title('Факт/предсказанное')
    plt.plot(target_test['2018-08-15':'2018-08-16'], label="факт")
    plt.plot(predicted['2018-08-15':'2018-08-16'], label="предсказаное")
    plt.legend()
    plt.show()
    
plotting(target_test, model.predict(features_test))
```

### CatBoostRegressor

```python
from catboost import CatBoostRegressor
model = CatBoostRegressor(iterations=50)
```

```python
model = CatBoostRegressor()
params = {'learning_rate' : [0.1, 0.15], 'depth': [1,5]}
split_index = TimeSeriesSplit().split(features_train)
```

```python
grid_search = GridSearchCV(model, params, scoring="neg_mean_squared_error", cv = split_index)
grid_search.fit(features_train, target_train, verbose=200)
```

```python
model=grid_search.best_estimator_
```

```python
grid_search.best_params_
```

```python
model.fit(features_train, target_train)
```

```python
print("RMSE:", mean_squared_error(model.predict(features_test), target_test)**0.5)
```

```python
plotting(target_test, model.predict(features_test))
```

**Вывод**    
Я попробовала обучить 2 модели: LinearRegression и CatBoostRegressor. Обе показали неплохой результат на тестовой выборке  RMSE< 48.  
Наилучший результат RMSE=39,47 показала модель CatBoostRegressor с гиперпараметрами depth=5 и learning_rate=0.1.   

## Общий вывод

Модель CatBoostRegressor с гиперпараметрами depth=5 и learning_rate=0.1 при проверке на тестовой выборке показала результат RMSE=39,47.
